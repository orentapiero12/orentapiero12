```python
import pandas as pd
```

```python
df = pd.read_csv('bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv')
df.index = pd.to_datetime(df['Timestamp'],unit = 's')
del df['Timestamp']
```

```python
df.Close.groupby(df.index.date).apply(lambda x: x.isna().sum())
```
